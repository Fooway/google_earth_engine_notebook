```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap
from IPython.core.display import Image
%matplotlib inline
```

```python
# lon_0 is central longitude of projection.
# resolution = 'c' means use crude resolution coastlines.
m = Basemap(projection='robin',lon_0=0,resolution='c')
m.drawcoastlines()
m.fillcontinents(color='bisque',lake_color='teal')
# draw parallels and meridians.
m.drawparallels(np.arange(-90.,120.,30.))
m.drawmeridians(np.arange(0.,360.,60.))
m.drawmapboundary(fill_color='teal')
plt.title("Robinson Projection")
plt.show()
```

```python
# practice showing a more local area
new_m = Basemap(projection='stere',
                lon_0=-72.5, 
                lat_0=42.5,
                llcrnrlon=-74.,llcrnrlat=40.,urcrnrlon=-70.,urcrnrlat=42.5,
                resolution='f') #decimal points on corner coordinates are necessary
new_m.drawstates()
new_m.drawcoastlines()
new_m.fillcontinents(color='bisque',lake_color='teal')
new_m.drawmapboundary(fill_color='teal')
plt.show()
```
